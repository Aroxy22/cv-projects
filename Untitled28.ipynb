Shape from Texture (3D Perception using Surface Patterns)

```python
!pip install --quiet opencv-python-headless==4.7.0.72 numpy matplotlib requests
```

```python
# Reliable brick wall image from Unsplash
url = "https://images.unsplash.com/photo-1523289333742-be1143f6b766?auto=format&w=800&fit=crop"
headers = {"User-Agent":"Mozilla/5.0"}
resp = requests.get(url, headers=headers, timeout=15)
img_array = np.asarray(bytearray(resp.content), dtype=np.uint8)
img_bgr = cv2.imdecode(img_array, cv2.IMREAD_COLOR)
if img_bgr is None:
    raise ValueError("Image failed to download")

img_bgr = cv2.resize(img_bgr, (400, 400))
gray = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)
cv2.imwrite(f"{RESULTS_DIR}/original.jpg", gray)

show_gray(gray, "Original Textured Surface ")
```

```python
# Compute local texture energy using variance in a sliding window
window_size = 15
mean = cv2.blur(gray.astype('float32'), (window_size, window_size))
sq_mean = cv2.blur((gray.astype('float32')**2), (window_size, window_size))
local_variance = sq_mean - mean**2

# Normalize to [0,1]
texture_energy = cv2.normalize(local_variance, None, 0, 1, cv2.NORM_MINMAX)

# Invert  regions with less texture (farther) get higher depth value
depth_map = 1 - texture_energy

cv2.imwrite(f"{RESULTS_DIR}/texture_energy.jpg", (texture_energy*255).astype(np.uint8))
cv2.imwrite(f"{RESULTS_DIR}/depth_map.jpg", (depth_map*255).astype(np.uint8))

# Display both maps
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(texture_energy, cmap='inferno')
plt.title("Texture Density Map")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(depth_map, cmap='terrain')
plt.title("Estimated Depth (Pseudo Shape from Texture)")
plt.axis('off')
plt.show()
```

```python
import json
summary = {
    "original_image": f"{RESULTS_DIR}/original.jpg",
    "texture_density_map": f"{RESULTS_DIR}/texture_energy.jpg",
    "pseudo_depth_map": f"{RESULTS_DIR}/depth_map.jpg",
    "method": "Local variance estimation (texture density -> depth cue)"
}
with open(f"{RESULTS_DIR}/summary.json", 'w') as f:
    json.dump(summary, f, indent=2)

!zip -r /content/shape_from_texture_results.zip {RESULTS_DIR} > /dev/null
from google.colab import files
files.download("/content/shape_from_texture_results.zip")
```

Optical Flow & Motion Estimation (Introduction to Motion)

```python
# Install libs (Colab usually has them)
!pip install --quiet opencv-python-headless==4.7.0.72 numpy matplotlib requests

import cv2, numpy as np, matplotlib.pyplot as plt, os, urllib.request, json
from matplotlib import cm
plt.rcParams['figure.dpi'] = 100

RESULTS_DIR = "/content/results_optical_flow"
os.makedirs(RESULTS_DIR, exist_ok=True)
```

```python
# Download a reliable sample video from OpenCV repo
video_url = "https://github.com/opencv/opencv/raw/master/samples/data/vtest.avi"
video_path = "/content/vtest.avi"
urllib.request.urlretrieve(video_url, video_path)

cap = cv2.VideoCapture(video_path)
# read frame 0 and frame 3 to create perceptible motion
ok, frame1 = cap.read()
cap.set(cv2.CAP_PROP_POS_FRAMES, 3)
ok2, frame2 = cap.read()
cap.release()

if frame1 is None or frame2 is None:
    raise RuntimeError("Failed to extract frames from video")

# Resize for speed & consistency
frame1 = cv2.resize(frame1, (640, 360))
frame2 = cv2.resize(frame2, (640, 360))

gray1 = cv2.cvtColor(frame1, cv2.COLOR_BGR2GRAY)
gray2 = cv2.cvtColor(frame2, cv2.COLOR_BGR2GRAY)

# display inputs
plt.figure(figsize=(10,4))
plt.subplot(1,2,1); plt.imshow(cv2.cvtColor(frame1, cv2.COLOR_BGR2RGB)); plt.title("Frame 1"); plt.axis('off')
plt.subplot(1,2,2); plt.imshow(cv2.cvtColor(frame2, cv2.COLOR_BGR2RGB)); plt.title("Frame 2"); plt.axis('off')
plt.show()

# save originals
cv2.imwrite(os.path.join(RESULTS_DIR, 'frame1.jpg'), frame1)
cv2.imwrite(os.path.join(RESULTS_DIR, 'frame2.jpg'), frame2)
```

```python
flow = cv2.calcOpticalFlowFarneback(
    gray1, gray2, None,
    pyr_scale=0.5, levels=3, winsize=15,
    iterations=3, poly_n=5, poly_sigma=1.2, flags=0
)

mag, ang = cv2.cartToPolar(flow[...,0], flow[...,1])
# HSV visualization
hsv = np.zeros_like(frame1)
hsv[...,1] = 255
hsv[...,0] = ang * 180 / np.pi / 2
hsv[...,2] = cv2.normalize(mag, None, 0, 255, cv2.NORM_MINMAX)
flow_rgb = cv2.cvtColor(hsv.astype(np.uint8), cv2.COLOR_HSV2RGB)

plt.figure(figsize=(10,5))
plt.imshow(flow_rgb); plt.title("Dense Optical Flow (Farneback)"); plt.axis('off')
plt.show()

cv2.imwrite(os.path.join(RESULTS_DIR, 'flow_rgb.jpg'), cv2.cvtColor(flow_rgb, cv2.COLOR_RGB2BGR))
```

```python
# prepare grid sample
step = 16  # spacing of arrows
h, w = gray1.shape
Y, X = np.mgrid[step/2:h:step, step/2:w:step].astype(int)
u = flow[Y, X, 0]
v = flow[Y, X, 1]

# build quiver visualization on top of frame2
plt.figure(figsize=(10,5))
plt.imshow(cv2.cvtColor(frame2, cv2.COLOR_BGR2RGB))
plt.quiver(X, Y, u, -v, color='red', angles='xy', scale_units='xy', scale=1.5, width=0.0025)
plt.title('Quiver: Sampled Motion Vectors'); plt.axis('off')
plt.show()

# Save the quiver overlay as image (draw arrows on image using cv2 for saving)
overlay = frame2.copy()
for (x,y,dx,dy) in zip(X.flatten(), Y.flatten(), u.flatten(), v.flatten()):
    x2 = int(x + dx*5)
    y2 = int(y + dy*5)
    cv2.arrowedLine(overlay, (int(x), int(y)), (x2,y2), color=(255,0,0), thickness=1, tipLength=0.3)
cv2.imwrite(os.path.join(RESULTS_DIR, 'flow_quiver_overlay.jpg'), overlay)

```

```python
# motion mask by thresholding magnitude
mag_norm = cv2.normalize(mag, None, 0, 255, cv2.NORM_MINMAX).astype(np.uint8)
_, motion_mask = cv2.threshold(mag_norm, 25, 255, cv2.THRESH_BINARY)  # tune threshold if needed

# morphological clean
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (7,7))
motion_mask = cv2.morphologyEx(motion_mask, cv2.MORPH_CLOSE, kernel, iterations=2)
motion_mask = cv2.morphologyEx(motion_mask, cv2.MORPH_OPEN, kernel, iterations=1)

# find contours -> bounding boxes
contours, _ = cv2.findContours(motion_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
bboxes = []
overlay_boxes = frame2.copy()
for cnt in contours:
    area = cv2.contourArea(cnt)
    if area < 300:  # filter small areas
        continue
    x,y,wc,hc = cv2.boundingRect(cnt)
    bboxes.append([int(x),int(y),int(wc),int(hc)])
    cv2.rectangle(overlay_boxes, (x,y), (x+wc, y+hc), (0,255,0), 2)

plt.figure(figsize=(10,5))
plt.subplot(1,2,1); plt.imshow(motion_mask, cmap='gray'); plt.title('Motion Mask'); plt.axis('off')
plt.subplot(1,2,2); plt.imshow(cv2.cvtColor(overlay_boxes, cv2.COLOR_BGR2RGB)); plt.title('Bounding Boxes on Moving Regions'); plt.axis('off')
plt.show()

cv2.imwrite(os.path.join(RESULTS_DIR, 'motion_mask.jpg'), motion_mask)
cv2.imwrite(os.path.join(RESULTS_DIR, 'motion_bboxes.jpg'), overlay_boxes)
```

```python
# detect good features in frame1
feature_params = dict(maxCorners=200, qualityLevel=0.01, minDistance=7, blockSize=7)
p0 = cv2.goodFeaturesToTrack(gray1, mask=None, **feature_params)
if p0 is None:
    p0 = np.empty((0,1,2), dtype=np.float32)

# LK params
lk_params = dict(winSize=(21,21), maxLevel=3,
                 criteria=(cv2.TERM_CRITERIA_EPS | cv2.TERM_CRITERIA_COUNT, 30, 0.01))

p1, st, err = cv2.calcOpticalFlowPyrLK(gray1, gray2, p0, None, **lk_params)
# select good points
good_old = p0[st==1]
good_new = p1[st==1]

sparse_overlay = frame2.copy()
for (old, new) in zip(good_old.reshape(-1,2), good_new.reshape(-1,2)):
    x0,y0 = old.ravel().astype(int)
    x1,y1 = new.ravel().astype(int)
    cv2.circle(sparse_overlay, (x1,y1), 3, (0,255,0), -1)
    cv2.line(sparse_overlay, (x0,y0), (x1,y1), (0,255,0), 1)

plt.figure(figsize=(10,5))
plt.imshow(cv2.cvtColor(sparse_overlay, cv2.COLOR_BGR2RGB))
plt.title(f'Sparse Tracking: {len(good_new)} features tracked')
plt.axis('off')
plt.show()

cv2.imwrite(os.path.join(RESULTS_DIR, 'sparse_tracking_overlay.jpg'), sparse_overlay)
```

```python
summary = {
    "dense_flow_image": os.path.join(RESULTS_DIR, 'flow_rgb.jpg'),
    "quiver_overlay": os.path.join(RESULTS_DIR, 'flow_quiver_overlay.jpg'),
    "motion_mask": os.path.join(RESULTS_DIR, 'motion_mask.jpg'),
    "motion_bboxes": os.path.join(RESULTS_DIR, 'motion_bboxes.jpg'),
    "sparse_tracking": os.path.join(RESULTS_DIR, 'sparse_tracking_overlay.jpg'),
    "motion_stats": {
        "mean_magnitude": float(np.mean(mag)),
        "max_magnitude": float(np.max(mag)),
        "num_moving_regions": len(bboxes),
        "num_tracked_keypoints": int(len(good_new))
    },
    "notes": "Dense + sparse optical flow, motion segmentation by mag threshold, bounding boxes for moving regions."
}

with open(os.path.join(RESULTS_DIR, 'summary.json'), 'w') as f:
    json.dump(summary, f, indent=2)

!zip -r /content/results_optical_flow.zip {RESULTS_DIR} > /dev/null
from google.colab import files
files.download('/content/results_optical_flow.zip')
print("Saved all outputs to /content/results_optical_flow/")
```

Principal Component Analysis (PCA) to extract key facial featurescalled Eigenfaces

```python
!pip install --quiet opencv-python-headless==4.7.0.72 numpy matplotlib scikit-learn
import numpy as np, matplotlib.pyplot as plt, cv2, os, json
from sklearn.decomposition import PCA
from sklearn.datasets import fetch_olivetti_faces
from sklearn.metrics import mean_squared_error
plt.rcParams['figure.dpi']=100

RESULTS_DIR="/content/results_eigenfaces"
os.makedirs(RESULTS_DIR, exist_ok=True)
```

```python
faces = fetch_olivetti_faces(shuffle=True, random_state=42)
X = faces.data   # (400, 4096)
images = faces.images
h, w = images[0].shape
print("Dataset:", X.shape, " (400 faces, each", h, "x", w, ")")

plt.figure(figsize=(8,4))
for i in range(8):
    plt.subplot(2,4,i+1)
    plt.imshow(images[i], cmap='gray')
    plt.axis('off')
plt.suptitle("Sample Faces from Olivetti Dataset")
plt.show()
```

```python
n_components = 25
pca = PCA(n_components=n_components, whiten=True).fit(X)
eigenfaces = pca.components_.reshape((n_components, h, w))

plt.figure(figsize=(8,6))
for i in range(10):
    plt.subplot(2,5,i+1)
    plt.imshow(eigenfaces[i], cmap='gray')
    plt.axis('off')
plt.suptitle("Top 10 Eigenfaces (Principal Components)")
plt.show()

# Save eigenface montage
eigenface_path=os.path.join(RESULTS_DIR,"eigenfaces.png")
plt.savefig(eigenface_path)
```

```python
X_pca = pca.transform(X)
reconstructed = pca.inverse_transform(X_pca)

# Show originals vs reconstructed
plt.figure(figsize=(10,5))
for i in range(5):
    plt.subplot(2,5,i+1)
    plt.imshow(images[i], cmap='gray')
    plt.title("Original"); plt.axis('off')
    plt.subplot(2,5,i+6)
    plt.imshow(reconstructed[i].reshape(h,w), cmap='gray')
    plt.title("Reconstructed"); plt.axis('off')
plt.suptitle("Top: Original | Bottom: Reconstructed using 25 Eigenfaces")
plt.show()

cv2.imwrite(os.path.join(RESULTS_DIR,"reconstruction_grid.jpg"),
            cv2.cvtColor((np.hstack([images[0],reconstructed[0].reshape(h,w)])*255).astype('uint8'),cv2.COLOR_GRAY2BGR))
```

```python
def recognize_face(idx_query, X, X_pca, pca):
    query_vec = X[idx_query]
    query_proj = X_pca[idx_query]
    distances = np.linalg.norm(X_pca - query_proj, axis=1)
    best_match = np.argmin(distances)
    return best_match, distances[best_match]

query_idx = 10
match_idx, dist = recognize_face(query_idx, X, X_pca, pca)

print(f"Best match for face {query_idx} is {match_idx} with distance {dist:.3f}")

plt.figure(figsize=(6,3))
plt.subplot(1,2,1); plt.imshow(images[query_idx], cmap='gray'); plt.title("Query Face"); plt.axis('off')
plt.subplot(1,2,2); plt.imshow(images[match_idx], cmap='gray'); plt.title("Matched Face"); plt.axis('off')
plt.show()

mse = mean_squared_error(X[query_idx], reconstructed[query_idx])
print(f"Reconstruction MSE: {mse:.5f}")
```

```python
summary={
    "num_components":n_components,
    "mean_reconstruction_error":float(np.mean((X-reconstructed)**2)),
    "example_query":query_idx,
    "best_match":int(match_idx),
    "distance":float(dist)
}
with open(os.path.join(RESULTS_DIR,"summary.json"),"w") as f:
    json.dump(summary,f,indent=2)

!zip -r /content/results_eigenfaces.zip {RESULTS_DIR} > /dev/null
from google.colab import files
files.download("/content/results_eigenfaces.zip")
```
